```python
from PIL import Image
import matplotlib.pyplot as plt
```

# Varying Height

```python
image = Image.open(r"tiff_files\Varying_Height\scalib_image.tif")
```

```python
plt.figure(figsize=(16, 10))
plt.imshow(image, cmap="gray")
plt.axvline(600)
plt.axvline(1200)
```

```python
pixel1 = (10.8-3.9)/600
pixel1
```

```python
plt.figure(figsize=(16, 10))
plt.imshow(image, cmap="gray")
plt.axvline(695)
plt.axvline(1103)
```

```python
pixel2 = (9.7-5)/(1103-695)
pixel2
```

```python
plt.figure(figsize=(16, 10))
plt.imshow(image, cmap="gray")
plt.axvline(800)
plt.axvline(1000)
```

```python
pixel3 = (8.5-6.2)/(1000-800)
pixel3
```

```python
fps = 1000
pixels = (pixel1+pixel2+pixel3)/3
pixels
```

The value of pixel per cm is **0.01151**.

```python
image.size[0]*pixels, image.size[1]*pixels
```
