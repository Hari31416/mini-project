```python
def get_num(value):
    reg = re.compile(r'\d+')
    try:
        num = reg.findall(value)[0]
    except:
        return str(99)
    if len(str(num)) == 1:
        num = "0"+num
    return num
```

```python
raw_df["Temp_Col"] = raw_df["ID"].apply(get_num)
smooth_df["Temp_Col"] = smooth_df["ID"].apply(get_num)
```

```python
raw_df = raw_df.sort_values(by="Temp_Col").reset_index(drop=True)
smooth_df = smooth_df.sort_values(by="Temp_Col").reset_index(drop=True)
```

```python
smooth_df.drop(columns=["Temp_Col"], inplace=True)
raw_df.drop(columns=["Temp_Col"], inplace=True)
```

```python
smooth_df.to_csv("results/info_on_smooth_data_all.csv", index=False)
raw_df.to_csv("results/info_on_raw_data_all.csv", index=False)
```
