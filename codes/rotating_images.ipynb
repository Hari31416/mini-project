## Rotation Using Transformation Matrix

The clockwise rotation matrix is given by:
$$
M(\theta) = \begin{bmatrix} cos\theta & -sin\theta \\ \\sin\theta& cos\theta \end{bmatrix}
$$
The inverse of the rotation matrix is given by:
$$
M^{-1}(\theta) = \begin{bmatrix} cos\theta & sin\theta \\ \\ -sin\theta& cos\theta \end{bmatrix}
$$

Which is nothing but the previous matrix with $-\theta$ instead of $+\theta$.

```python
def rotate_matrix(A, theta):
    theta = np.deg2rad(theta)
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta), np.cos(theta)]])
    return np.dot(R, A)
```
