```python
    def extract_image(image_path,crop=True,plot1=True,plot2=True):
        image_path = 'Image.jpg'
        im = plt.imread(image_path)
        if crop:
            image_arr = im[400:500,800:1000]
            #plt.imshow(image_arr,cmap='gray')
        else:
            image_arr=im
            #plt.imshow(image_arr,cmap='gray')
            
        fin_im=image_arr>110
        #plt.imshow(fin_im,cmap='gray')
        xs = fin_im.argmin(axis=0)
        print('xs',xs)
        print('xs-shape:',xs.shape)
        xss = np.nonzero(xs)
        print('xss',xss)
        xl=xss[0][0]
        xr=xss[0][-1]

        ys = fin_im.argmin(axis=1)
        print('ys',ys)
        print('ys-shape:',ys.shape)
        yss = np.nonzero(ys)
        print('yss',yss)
        yu=yss[0][0]
        yl=yss[0][-1]
        r1 = np.abs((xl-xr)/2)
        r2 = np.abs((yu-yl)/2)
        r = int((r1+r2)/2)
        xc = int(800+xl+r)
        yc=int(400+yu+r)

        if plot1:
            plt.figure(figsize=(8,6))
            plt.imshow(fin_im,cmap='gray')
            plt.hlines(y =yu, xmin = 30, xmax = 100)
            plt.hlines(y = yl, xmin = 30, xmax = 100)
            plt.hlines(y = yu+abs((yl-yu)/2), xmin = 30, xmax = 100,color='g')
            plt.vlines(x=xl,ymin=40,ymax=85,color='r')
            plt.vlines(x=xr,ymin=40,ymax=85,color='r')
            plt.vlines(x=xl+(abs(xl-xr)/2),ymin=40,ymax=85,color='g')


        if plot2:
            plt.figure(figsize=(8,6))
            plt.imshow(im,cmap='gray')
            plt.axhline(y=yc)
            plt.axvline(x=xc)
            plt.annotate((xc,yc),xy=(xc,yc) ,xytext=(xc+8,yc+8),color='r')
            
            
        print('xc,yc,r')
        return (xc,yc,r)


            
```
